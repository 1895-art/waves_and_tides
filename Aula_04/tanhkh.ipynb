```python
from __future__ import division

import numpy as np
import matplotlib.pyplot as plt
```

```python
# From zero up to a unrealistic wave where h == L.
hoL = np.arange(0, 1, 0.001)
kh = 2 * np.pi * hoL
tanhkh = np.tanh(kh)

# Get kh << 1 and kh == 1
threshold = 1e-3
mask = np.logical_and((1 - tanhkh) > threshold, (kh - tanhkh) > threshold)

fig, ax = plt.subplots(figsize=(8, 4))
label=r"$\tanh(kh)$ << 1 or $\tanh(kh) \sim kh$"
ax.plot(hoL, tanhkh, 'black', linewidth='3', label=label)
ax.plot(hoL[mask], tanhkh[mask], 'gray', linewidth='3')
ax.set_xlabel(r'$\frac{h}{L}$')
ax.set_ylabel(r'$\tanh(kh)$')
ax.axis([0, 1, 0, 1.05])
ax.legend(loc='lower right')

plt.show()
```
