```python
from __future__ import division

import numpy as np
import matplotlib.pyplot as plt

x0, y0 =1, 1
r = -0.000001
t = np.arange(1, 10000000)
f0 = -0.0001

x = (x0 / r) * (np.exp(r * t)) * (f0 * np.sin(f0 * t) + r * np.cos(f0 * t))
y = -(y0 / f0) * (np.exp(r * t)) * (r * np.sin(f0 * t) - f0 * np.cos(f0 * t))

fig, ax = plt.subplots()
ax.plot(x, y)
```
